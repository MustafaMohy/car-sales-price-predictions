# Read Data

```python
df = pd.read_csv('/kaggle/input/ann-car-sales-price-prediction/car_purchasing.csv',encoding='ISO-8859-1')
df.head()
```

**Drop unneeded columns** 

```python
df = df.drop(['customer name','customer e-mail','country','gender'], axis=1)
```

```python
df.head(3)
```

Preprocessing

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = df.iloc[:,:-1]
y =df.iloc[:,-1]
```

```python
scaler.fit(x)
```

```python
x = scaler.transform(x)
```

# Spliting Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train, y_test = train_test_split(x, y, test_size =0.2 , random_state =42)
```

```python
print(f"x_train.shape: {X_train.shape}")
print(f"x_test.shape: {X_test.shape}")
print("_____________\n")
print(f"y_train.shape: {y_train.shape}")
print(f"y_test.shape: {y_test.shape}")
```

# Apply Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
linear=reg.predict(X_test)
print(f"reg.score for test: {reg.score(X_test,y_test)}")
print(f"reg.score for train: {reg.score(X_train,y_train)}")
```
